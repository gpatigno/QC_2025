```python
import matplotlib.pyplot as plt

from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, transpile
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_distribution, plot_histogram, array_to_latex
```

```python
# Simulador para simular el siguiente circuito cu√°ntico.
simulator = AerSimulator()
```

```python
q = QuantumRegister(1, name='q')
c = ClassicalRegister(1, name='c')
```

### Compuerta de Hadamard:

```python
qc = QuantumCircuit(q, c)
# Se inicia el circuito en el estado |0>:
qc.reset(q[0])
# Aplica la compuerta de Hadamard:
qc.h(q)

# Se visualiza el circuito resultante:
qc.draw('mpl')
```

```python
HadamardCircuit = qc.copy()
VectorEstado = Statevector(HadamardCircuit)
array_to_latex(VectorEstado, prefix="\\ket{\\psi_1} = ")
```

```python
HadamardCircuit.measure(q, c)

# Se visualiza el circuito resultante:
HadamardCircuit.draw('mpl')
```

```python
job = simulator.run(transpile(HadamardCircuit, simulator), shots=1024)
counts = job.result().get_counts(HadamardCircuit)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Plot histogram in each subplot
plot_histogram(counts, ax=axs[0])
plot_distribution(counts, ax=axs[1])

axs[0].set_title("Frecuencias de resultados")
axs[1].set_title("Probabilidades basadas en las frecuencias")

plt.tight_layout()
plt.show()
```

Aqu√≠, la puerta de Hadamard crea una superposici√≥n igual de los estados $\left|0\right\rangle$ y $\left|1\right\rangle$, cada uno con una probabilidad del 50%.
